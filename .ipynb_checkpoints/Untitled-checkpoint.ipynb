```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
def graph(y_actual, y_pred, buy_prices, pnl, start = None, end = None, tick = 0):

    if not start:
        start = 0
    if not end:
        end = len(y_actual)

    
    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(len(buy_prices))[start:end], buy_prices[start:end], label="Buy", color='green', linewidth=2)
    plt.plot(np.arange(len(y_actual))[start:end], y_actual[start:end], label="Actual", color='blue', linewidth=2)
    plt.plot(np.arange(len(y_pred))[start:end], y_pred[start:end], label="Predicted", color='red', linestyle='dashed', linewidth=2)
    plt.xlabel("Time Steps")
    plt.ylabel("Value")
    if tick:
        plt.xticks(np.arange(start, end, tick))
    plt.title("Actual vs Predicted Over Time")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(len(pnl))[start:end], pnl[start:end], label="PNL", color='blue', linewidth=2)
    plt.xlabel("Time Steps")
    plt.ylabel("Value")
    plt.title("PnL")
    
    plt.legend()
    plt.grid(True)
    plt.show()
    
```

```python
trades = pd.read_csv("simulations/BTC-USD_2025-03-02 22:31:44.474315_40_update_linear_regressor.pkl.csv")
actual_prices = trades["sell_price"]
predicted_prices = trades["predicted_sell_price"]
buy_prices = trades["buy_price"]
pnl = trades["pnl"]
graph(actual_prices, predicted_prices, buy_prices, pnl)

```
