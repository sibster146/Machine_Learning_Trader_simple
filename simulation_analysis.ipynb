```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
simulation = pd.read_csv("simulations/simulation_11.csv")
simulation["Change in Price"] = simulation["buy_price"] - simulation.iloc[0]["buy_price"]
plt.figure(figsize=(10, 5))
plt.scatter(simulation["sell_sequence_number"], simulation["pnl"], label="PNL", color='blue', s = 1)
plt.plot(simulation["buy_sequence_number"], simulation["Change in Price"], label="Change in Asset Price", color='red', linestyle='dashed', linewidth=2)
plt.xlabel("Sequence Number")
plt.ylabel("Change in Value ($)")
plt.title("PNL vs Change in Asset Mid Price")
plt.legend()
plt.grid(True)
plt.show()
```
